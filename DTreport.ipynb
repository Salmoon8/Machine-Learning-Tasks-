```python
import numpy as np
from collections import Counter
from statistics import mode
from sklearn.metrics import accuracy_score

class Question:


    """Questions are used to partition a dataset into subtables according to discrete values."""
    def __init__(self, column, partition_values):
        self.column = column  # the feature we are splitting on
        self.partition_values = partition_values  # the unique/discrete values of the node question(branches)


    def match(self, row):
        """matches row feature with the  idx of child branch it should go to."""
        """" suppose we have a feat with low, mid, high as values during classification the match function
        returns the idx of the child branch we move to according to row column value """



        for t, val in enumerate(self.partition_values):
            if row[self.column] == val:
                return t

        # if it does not match with any data seen on training return 0 (first child)
        return 0


class node:
    """ Holds a reference to the question, and the child nodes that is partitioned by the question. """
    def __init__(self, child_branches, question):
        self.child_branches = child_branches
        self.question = question













```

```python
class Leaf:
    """
    holds labels for pure nodes

    used in classification when traversing the tree
    """
    def __init__(self, most_common):

        self.label=most_common
```

```python
class id3:
    def __init__(self, min_samples_split=2, max_depth=5):
        self.min_samples_split=min_samples_split
        self.max_depth=max_depth
        self.root=None

    def fit(self, X, y):
                self.root = self._build_tree(X, y, 0, [])

    def class_counts(self,numpy_arr):
        """Returns the unique elements that are inside the
        numpy array, and their individual counts"""

        if len(numpy_arr.shape)==2:
            valueCounts = pd.Series(numpy_arr[:, 0]).value_counts()
            return np.array(valueCounts.keys().to_list()), np.array(valueCounts.to_list())
        else:
            valueCounts = pd.Series(numpy_arr).value_counts()
        return np.array(valueCounts.keys().to_list()), np.array(valueCounts.to_list())

    def find_best_split(self, X, y, used_attributes):
        """
        loop opver every feature (except features used in a previous split).
        Choose the best possible attribute and best possible split in terms of information gain.
        """
        copy_used_attributes = used_attributes[0:]
        best_gain = 0
        best_question = None
        best_partition = []
        used_attribute = 0  # To not use any attribute (to split the data-subset) twice, in a single path from root to leaf.
        current_entropy = self.entropy(y)
        n_features = X.shape[1]


        for col in range(n_features):
            if col in copy_used_attributes:
                continue

            # multi-way split: as many branches as len(unique_values)
            unique_values, counts = self.class_counts(X[:,col])
            partitions_to_indexesOfSamples = []

            for value in unique_values:
                indexes_in_this_partition = X[:, col] == value
                partitions_to_indexesOfSamples.append(indexes_in_this_partition)

            # calculate info gain after getting idxs of feature values

            gain = self.information_gain(y, partitions_to_indexesOfSamples, current_entropy)



            if gain >= best_gain:
                best_gain = gain
                best_partition = partitions_to_indexesOfSamples
                best_question = Question(col, unique_values)
                used_attribute = col

        copy_used_attributes.append(used_attribute)
        return best_gain, best_partition, best_question, copy_used_attributes



    def _build_tree(self, X_train, y_train, depth, used_attributes):
        """ recursive decision tree building function """

        # base cases // stopping criterias

        if X_train.shape[0] < self.min_samples_split:
            return Leaf(self._most_common_label(y_train))

        if depth >self.max_depth:
            return Leaf(self._most_common_label(y_train))



        gain, partitions, question, used_attributes = self.find_best_split(X_train, y_train, used_attributes)

        if gain == 0 or question is None:
            return Leaf(self._most_common_label(y_train)) # no gain from splitting or we used all features


        child_branches = []
        for indices_in_this_partition in partitions:
            child_branches.append( self._build_tree(X_train[indices_in_this_partition], y_train[indices_in_this_partition], depth+1, used_attributes) )

        return node(child_branches, question)


    def entropy(self,y):
        n=y.shape[0]
        classes,counts=self.class_counts(y)
        p=np.true_divide(counts,n)
        entropy=-np.sum(np.multiply(p,np.log2(p)))

        return entropy

    def information_gain(self,y, intervals_indexes, current_entropy):
        """Information Gain:  The total Entropy of the parent node, minus the weighted entropy of child nodes. """
        sum_ = 0
        len_sum = 0

        # weighted sum of entropies of each split (or interval)
        for interval_indices in intervals_indexes:
            sum_ += float(len(interval_indices) * self.entropy(y[interval_indices]))
            len_sum += len(interval_indices)

        return current_entropy - sum_ / len_sum


    def _split(self, column):
        uniquevalues,counts=self.class_counts(column)
        indices=[]
        for x in uniquevalues:
            idx=np.argwhere(column==x).flatten()
            indices.append(idx)

        return indices

    def _most_common_label(self, y):
        flaty = [element for array in y for element in array]
        return max(flaty, key = flaty.count)

    def predict(self, X_test):
        preds = []

        for i, row in enumerate(X_test):
            preds.append(self._classify(row,self.root))

        preds = np.array(preds)


        return preds

    def _classify(self, row, node):
        """Recursively traverses the decision tree towards a prediction(Leaf node)"""

        # Base case: At a leaf node
        if isinstance(node, Leaf):
            return node.label


        branch_to_go = node.child_branches[node.question.match(row)]
        return self._classify(row, branch_to_go)



```
