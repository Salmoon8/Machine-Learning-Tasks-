```python
import pandas as pd
import numpy as np
```

## First let's start by reading the data and selecting only the first 200 users


```python
ratings=pd.read_csv('ratings.csv')
```

```python
movies=pd.read_csv('movies.csv')
```

```python
ratings_200=ratings[ratings.userId<=200]
movieids=np.array(ratings_200.movieId)
```

## Filtering the movies to those from the first 200 users only

```python
movies_200=movies.loc[movies.movieId.isin (movieids) ]

```

```python
ratings_200
```

```python
movies_200
```

### Merge the two tables on movieId

```python
data=ratings_200.merge(movies_200,on='movieId')
```

```python
data
```

### Now we will make a pibot table based on user ratings and movies only and fill movie that the user has not watched with 0

```python
pivot_table_user = pd.pivot_table(data=data,values='rating',index='userId',columns='title')
pivot_table_user


```

### now we will subtract the mean from the ratings so the data is centered around 0 (clearer ratings) and not affected by the movies that the user has not watched as the mean will be zero

```python
user_ratings_mean = np.nanmean(pivot_table_user,axis=1 )
user_ratings_mean.reshape(-1,1)

demeaned_pivot = pivot_table_user - user_ratings_mean.reshape(-1,1)
demeaned_pivot=demeaned_pivot.fillna(0)
demeaned_pivot


```

### Now we will compute the similarity matrix according to the cosine similarity function

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(demeaned_pivot.T)
cosine_sim.shape
```

```python
similarity_df1 = pd.DataFrame(cosine_sim, columns=demeaned_pivot.columns, index=demeaned_pivot.columns)
```

```python
similarity_df1
```

### now create a function to get the 10 most similar movies according to the similarity matrix score

```python
def get_similar_movies(movie_title, top_n=10):
    # Get the similarity scores for the given movie
    similarity_scores = similarity_df1[movie_title]

    # Sort the similarity scores in descending order and select the top N similar movies
    similar_movies = similarity_scores.sort_values(ascending=False).index[:top_n]


    return similar_movies
```

```python
similar_movies_1 = get_similar_movies("Toy Story (1995)"
                                     )
```

```python
similar_movies_1
```

```python
similar_movies_2 = get_similar_movies("Waiting to Exhale (1995)"
                                     )
```

```python
similar_movies_2
```
